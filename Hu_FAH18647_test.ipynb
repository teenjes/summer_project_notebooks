```python
lengths = []
for key in fastq_dict:
    lengths.append(len(fastq_dict[key].seq))
len(lengths)
```

```python
fastq_lengths_dataframe = df(data=["%.2f" % max(lengths), "%.2f" % min(lengths), "%.2f" % np.mean(lengths), 
                "%.2f" % np.median(lengths)], index=["max", "min", "mean", "median"], 
                columns=["Value"])
```

```python
fastq_qscores_dataframe = df(data=["%.2f" % max(fastq_dict['e212a37c-cda0-4984-8c10-bdbbdf99b370'].letter_annotations["phred_quality"]), "%.2f" % min(fastq_dict['e212a37c-cda0-4984-8c10-bdbbdf99b370'].letter_annotations["phred_quality"]), "%.2f" % np.mean(fastq_dict['e212a37c-cda0-4984-8c10-bdbbdf99b370'].letter_annotations["phred_quality"]), 
                "%.2f" % np.median(fastq_dict['e212a37c-cda0-4984-8c10-bdbbdf99b370'].letter_annotations["phred_quality"])], index=["max", "min", "mean", "median"], 
                columns=["Value"])
```

```python
fastq_qscores_dataframe
```

```python
mean_qscores = []
for key in fastq_dict:
    mean_qscores.append(np.mean(fastq_dict[key].letter_annotations["phred_quality"]))
```

```python
ax = sns.distplot(mean_qscores, color="k")
ax.set_xlabel("Q Score")
ax.set_ylabel("Frequency")
```
