```python
import Bio
from Bio import SeqIO
import numpy as np
import pandas
from pandas import DataFrame as df
import seaborn as sns
import pandas as pd
```

```python
fastq_5ug_fail = SeqIO.to_dict(SeqIO.parse("/media/MassStorage/tmp/TE/summer_project/raw_data/test_data/20171212_0416_121217_Dave_long_read_5ug_fail.fastq", "fastq"))
```

```python
len(fastq_5ug_fail)
```

```python
lengths = []
for key in fastq_5ug_fail:
    lengths.append(len(fastq_5ug_fail[key].seq))
```

```python
fastq_5ug_fail_lengths_dataframe = df(data=["%.2f" % max(lengths), "%.2f" % min(lengths), "%.2f" % np.mean(lengths), 
                "%.2f" % np.median(lengths)], index=["max", "min", "mean", "median"], 
                columns=["Value"])
```

```python
fastq_5ug_fail_lengths_dataframe
```

```python
ax = sns.distplot(lengths, color="k")
ax.set_title("5ug Fail Read Lengths")
ax.set_xlabel("Read Lengths")
ax.set_ylabel("Frequency")
```

```python
ax = sns.distplot(lengths, color="k", bins=10000)
ax.set_title("5ug Fail Read Lengths")
ax.set_xlabel("Read Lengths")
ax.set_ylabel("Frequency")
ax.set_xlim(0,4000)
```

```python
fastq_5ug_fail_mean_qscores = []
for key in fastq_5ug_fail:
    fastq_5ug_fail_mean_qscores.append(np.mean(fastq_5ug_fail[key].letter_annotations["phred_quality"]))
```

```python
ax = sns.distplot(fastq_5ug_fail_mean_qscores, color="k")
ax.set_title("5ug Fail Q Scores")
ax.set_xlabel("Q Score")
ax.set_ylabel("Frequency")
```

```python
ax = sns.jointplot(x=lengths, y=fastq_5ug_fail_mean_qscores, color="k")
```
